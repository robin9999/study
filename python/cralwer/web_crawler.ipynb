```python
# parser.py
import requests
from bs4 import BeautifulSoup

#한건의 대화를 담는 클래스를 정의

class Conversation:
    #question, answer  2 variable
    def __init__(self,question,answer):
        self.question = question
        self.answer = answer
        
    def __str__(self):
        return "질문 : " +self.question +"\n답변 : "+ self.answer +"\n"

#모든 영어 대화 주제 추출
def get_subjects():
    subjects=[]
    
    # 전체 주제목록을 보여주는 페이지로의 요청(request) 객체를 생성.
    req = requests.get('https://basicenglishspeaking.com/daily-english-conversation-topics/')
    html = req.text
    soup = BeautifulSoup(html,'html.parser')
    
    divs=soup.findAll('div', {"class" :"su-column-inner su-clearfix"})
    
    for div in divs:
        # 내부에 존재하는 a태그들만 추출
        links = div.findAll('a')
        
        #  <a> 태그  내부의 텍스트를  리스트에 삽입한다.
        for link in links:
            subject = link.text
            subjects.append(subject)
            
    return subjects
    
subjects = get_subjects()
    
print('총 : ',len(subjects), '개의 주제를 찾았습니다.')
print(subjects)
    
```

```python
conversations =[]

i =1 

#모든 대화 주제 각각에 접근
for sub in subjects:
    print( '(' , i , ' / ' , len(subjects),' )',sub )
    #대화 스크립트 보여주는 페이지로의 요청(Request) 객체 생성
    req = requests.get('https://basicenglishspeaking.com/family/' + sub)
    html = req.text
    
    soup = BeautifulSoup(html,'html.parser')
    qnas = soup.findAll('div', {"class": "sc_player_container1"})
    
    #각각의 대화내용에 모두 접근한다.
    for qna in qnas:
        if qnas.index(qna)%2 ==0:
            q = qna.next_sibling
        else:
            a = qna.next_sibling
            c=Conversation(q,a)
            conversations.append(c)
        
        i = i+1
        
        if i ==10:
            break;
        
    print('총 ',len(conversations),'개의 대화를 찾았습니다.')
    
    for c in conversations:
        print(str(c))
```
